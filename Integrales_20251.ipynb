```r
# Cargar librería
library(ggplot2)

# 1. Definir la función
f <- function(x) x

# 2. Intervalo
a <- 0
b <- 1

# 3. Calcular la integral exacta
area_exacta <- integrate(f, lower = a, upper = b)$value

# 4. Crear un data frame para graficar f(x) = x
x_vals <- seq(a, b, length.out = 100)
df_func <- data.frame(x = x_vals, y = f(x_vals))

# 5. Crear data para sombrear el área bajo la curva
df_area <- data.frame(x = c(a, x_vals, b),
                      y = c(0, f(x_vals), 0))

# 6. Graficar
ggplot() +
  geom_line(data = df_func, aes(x = x, y = y), color = "blue", size = 1.2) +
  geom_polygon(data = df_area, aes(x = x, y = y), fill = "skyblue", alpha = 0.5) +
  labs(
    title = "Área bajo la curva de f(x) = x en [0, 1]",
    subtitle = paste("Área exacta =", round(area_exacta, 4)),
    x = "x",
    y = "f(x)"
  ) +
  theme_minimal()
```

```r
library(ggplot2)

# ============================
# EJEMPLO 6
# ============================
#(3/14)*
# Definir función f6
f6 <- function(x) ifelse(x >= 0 & x <= 5, (2*x^2 + 2*x + 3), 0)

# Valores para graficar
x6 <- seq(-0.5, 2.5, length.out = 500)
y6 <- f6(x6)
df6 <- data.frame(x = x6, y = y6)

# Gráfico con ggplot2
ggplot(df6, aes(x = x, y = y)) +
  geom_line(color = "blue", size = 1.2) +
  geom_area(data = subset(df6, y > 0), fill = "skyblue", alpha = 0.5) +
  labs(title = "Ejemplo 6: f(x) = (3/14)(2x^2 + 2x + 3) en [0,1]",
       x = "x", y = "f(x)") +
  theme_minimal()

# Calcular el valor de la integral en [0, 10]
area <- integrate(f6, lower = 0, upper = 10)$value
cat("Área bajo la curva en [0,10] =", area, "\n")

# ============================
# Cálculos: Área, Media y Varianza
# ============================

# Área bajo la curva
area_total <- integrate(f6, lower = 0, upper = 1)$value
cat("Área total bajo la curva =", area_total, "\n")

# Media (valor esperado)
f6_media <- function(x) x * f6(x)
media <- integrate(f6_media, lower = 0, upper = 1)$value
cat("Media E[X] =", media, "\n")

# E[X^2]
f6_x2 <- function(x) x^2 * f6(x)
esperado_x2 <- integrate(f6_x2, lower = 0, upper = 1)$value

# Varianza = E[X^2] - (E[X])^2
varianza <- esperado_x2 - media^2
cat("Varianza Var(X) =", varianza, "\n")
```

```r
# ---------------------------------------------------------------
# PRUEBA DE LOS SIGNOS Y WILCOXON - MEJORA EN GALVANIZADO
# ---------------------------------------------------------------

# Cargar librerías necesarias
library(ggplot2)

# -------------------------------
# 1. Datos
# -------------------------------
antiguo <- c(10.2, 9.8, 10.1, 9.7, 10.5, 10.0, 10.3, 10.2)
nuevo   <- c(10.6, 9.9, 10.0, 10.1, 10.7, 10.0, 10.4, 10.6)

# -------------------------------
# 2. Prueba de los Signos
# -------------------------------

# Calcular diferencias
diferencias <- nuevo - antiguo

# Eliminar las diferencias iguales a cero
diferencias_sin_cero <- diferencias[diferencias != 0]

# Número total de observaciones útiles
n <- length(diferencias_sin_cero)

# Número de diferencias positivas
positivos <- sum(diferencias_sin_cero > 0)

# Cálculo del valor p (binomial unilateral derecha)
p_valor <- sum(dbinom(positivos:n, size = n, prob = 0.5))

# Mostrar resultados
cat("----- PRUEBA DE LOS SIGNOS -----\n")
cat("n =", n, "\n")
cat("Signos positivos =", positivos, "\n")
cat("Valor p =", round(p_valor, 4), "\n")
if(p_valor < 0.05){
  cat("Decisión: Rechazamos H0 → El nuevo proceso mejora el recubrimiento.\n")
} else {
  cat("Decisión: No se rechaza H0 → No hay mejora significativa.\n")
}

# -------------------------------
# 3. Visualización con ggplot2
# -------------------------------

# Crear data frame
df <- data.frame(
  Pieza = factor(1:8),
  Diferencia = diferencias,
  Signo = ifelse(diferencias > 0, "Positiva",
                 ifelse(diferencias < 0, "Negativa", "Sin cambio"))
)

# Gráfico de diferencias
ggplot(df, aes(x = Pieza, y = Diferencia, fill = Signo)) +
  geom_bar(stat = "identity", color = "black") +
  geom_hline(yintercept = 0, linetype = "dashed") +
  scale_fill_manual(values = c("Positiva" = "steelblue", "Negativa" = "firebrick", "Sin cambio" = "gray")) +
  labs(
    title = "Diferencias entre método nuevo y antiguo",
    y = "Diferencia (micrones)",
    x = "Pieza"
  ) +
  theme_minimal()

# -------------------------------
# 4. Prueba de Wilcoxon
# -------------------------------
cat("\n----- PRUEBA DE WILCOXON (pares relacionados) -----\n")
wilcox_result <- wilcox.test(nuevo, antiguo, paired = TRUE, alternative = "greater")
print(wilcox_result)
```

```r
# ---------------------------------------------------------------
# PRUEBA DE LOS SIGNOS - TÉCNICA DE ENSEÑANZA
# ---------------------------------------------------------------

library(ggplot2)

# Datos
antes <- c(3.5, 3.8, 3.0, 4.1, 3.7, 2.8, 3.6, 4.0, 3.3, 3.5)
despues <- c(4.0, 4.2, 2.9, 4.3, 3.7, 3.2, 3.9, 4.2, 3.6, 3.2)

# Diferencias
diferencias <- despues - antes

# Quitar ceros
diferencias_sin_cero <- diferencias[diferencias != 0]
n <- length(diferencias_sin_cero)
positivos <- sum(diferencias_sin_cero > 0)

# Valor p
p_valor <- sum(dbinom(positivos:n, size = n, prob = 0.5))

# Resultados
cat("----- PRUEBA DE LOS SIGNOS -----\n")
cat("n =", n, "\nSignos positivos =", positivos, "\nValor p =", round(p_valor, 4), "\n")
if(p_valor < 0.05){
  cat("Decisión: Rechazamos H0 → La técnica mejora significativamente.\n")
} else {
  cat("Decisión: No se rechaza H0 → No hay mejora significativa.\n")
}

# Visualización
df <- data.frame(
  Estudiante = factor(1:10),
  Diferencia = diferencias,
  Signo = ifelse(diferencias > 0, "Positiva",
                 ifelse(diferencias < 0, "Negativa", "Sin cambio"))
)

ggplot(df, aes(x = Estudiante, y = Diferencia, fill = Signo)) +
  geom_bar(stat = "identity", color = "black") +
  geom_hline(yintercept = 0, linetype = "dashed") +
  scale_fill_manual(values = c("Positiva" = "steelblue", "Negativa" = "firebrick", "Sin cambio" = "gray")) +
  labs(
    title = "Diferencias en calificaciones con la nueva técnica",
    y = "Diferencia (Después - Antes)",
    x = "Estudiante"
  ) +
  theme_minimal()

# Prueba de Wilcoxon
cat("\n----- PRUEBA DE WILCOXON -----\n")
wilcox.test(despues, antes, paired = TRUE, alternative = "greater")
```

```r
# ---------------------------------------------------------------
# EJEMPLO 16.2 - PRUEBA DE LOS SIGNOS Y WILCOXON
# Comparación entre neumáticos radiales y con cinturón
# ---------------------------------------------------------------

# Cargar librería para graficar
library(ggplot2)

# -------------------------------
# 1. Datos
# -------------------------------
radial <- c(4.2, 4.7, 6.0, 6.7, 7.2, 7.0, 5.7, 6.0,
            7.4, 4.9, 6.9, 6.1, 5.2, 6.6, 7.4, 6.9)

cinturon <- c(4.1, 4.9, 5.9, 6.4, 6.8, 6.6, 5.9, 5.8,
              6.9, 4.9, 6.0, 5.6, 5.3, 7.1, 7.5, 6.4)

# -------------------------------
# 2. Prueba de los Signos
# -------------------------------

# Calcular diferencias
diferencias <- radial - cinturon

# Eliminar ceros
diferencias_sin_cero <- diferencias[diferencias != 0]
n <- length(diferencias_sin_cero)
positivos <- sum(diferencias_sin_cero > 0)

# Cálculo del valor p (prueba unilateral derecha)
p_valor <- sum(dbinom(positivos:n, size = n, prob = 0.5))

# Mostrar resultados
cat("----- PRUEBA DE LOS SIGNOS -----\n")
cat("Número de datos sin empate:", n, "\n")
cat("Número de signos positivos:", positivos, "\n")
cat("Valor p (unilateral derecha):", round(p_valor, 4), "\n")
if(p_valor < 0.05){
  cat("Decisión: Rechazamos H0 → Los neumáticos radiales ahorran más combustible.\n")
} else {
  cat("Decisión: No se rechaza H0 → No hay diferencia significativa.\n")
}

# -------------------------------
# 3. Visualización con ggplot2
# -------------------------------

# Preparar dataframe
df <- data.frame(
  Automovil = factor(1:16),
  Diferencia = diferencias,
  Signo = ifelse(diferencias > 0, "Positiva",
                 ifelse(diferencias < 0, "Negativa", "Sin cambio"))
)

# Gráfico
ggplot(df, aes(x = Automovil, y = Diferencia, fill = Signo)) +
  geom_bar(stat = "identity", color = "black") +
  geom_hline(yintercept = 0, linetype = "dashed") +
  scale_fill_manual(values = c("Positiva" = "steelblue",
                               "Negativa" = "firebrick",
                               "Sin cambio" = "gray")) +
  labs(
    title = "Diferencias en rendimiento de combustible",
    subtitle = "Neumáticos radiales - con cinturón (km/litro)",
    x = "Automóvil",
    y = "Diferencia",
    fill = "Signo"
  ) +
  theme_minimal()

# -------------------------------
# 4. Prueba de Wilcoxon
# -------------------------------

cat("\n----- PRUEBA DE WILCOXON (pares relacionados) -----\n")
resultado_wilcoxon <- wilcox.test(radial, cinturon, paired = TRUE, alternative = "greater")
print(resultado_wilcoxon)
```

# **Prueba de la Suma de Rangos de Wilcoxon (Mann-Whitney)**

Comparar si dos muestras independientes provienen de poblaciones con igual distribución central, sin asumir normalidad ni igualdad de varianzas.

```r
# Datos del ejemplo
linea_a <- c(20, 22, 24, 25, 23)
linea_b <- c(27, 30, 28, 32, 31)

# Prueba de Wilcoxon para muestras independientes (Mann-Whitney U)
wilcox.test(linea_a, linea_b, alternative = "two.sided", exact = TRUE)
```

```r
# ---------------------------------------------------------------
# PRUEBA DE SUMA DE RANGOS DE WILCOXON (MANN-WHITNEY)
# Ejemplo 16.5 - Comparación del contenido de nicotina
# ---------------------------------------------------------------

# Cargar librerías
library(ggplot2)

# -------------------------------
# 1. Datos
# -------------------------------
marca_a <- c(2.1, 4.0, 6.3, 5.4, 4.8, 3.7, 6.1, 3.3)
marca_b <- c(4.1, 0.6, 3.1, 2.5, 4.0, 6.2, 1.6, 2.2, 1.9, 5.4)

# -------------------------------
# 2. Prueba de Wilcoxon (Mann-Whitney U)
# -------------------------------
resultado <- wilcox.test(marca_a, marca_b, alternative = "two.sided", exact = TRUE)

cat("----- PRUEBA DE MANN-WHITNEY -----\n")
print(resultado)

# -------------------------------
# 3. Preparar datos para ggplot2
# -------------------------------
grupo_a <- data.frame(Nicotina = marca_a, Marca = "A")
grupo_b <- data.frame(Nicotina = marca_b, Marca = "B")
datos <- rbind(grupo_a, grupo_b)

# -------------------------------
# 4. Visualización con ggplot2
# -------------------------------
ggplot(datos, aes(x = Marca, y = Nicotina, fill = Marca)) +
  geom_boxplot(width = 0.4, outlier.color = "black") +
  geom_jitter(width = 0.1, alpha = 0.7) +
  labs(
    title = "Comparación del contenido de nicotina por marca",
    subtitle = "Ejemplo 16.5 - Prueba de suma de rangos de Wilcoxon",
    x = "Marca",
    y = "Contenido de nicotina (mg)"
  ) +
  scale_fill_manual(values = c("A" = "skyblue", "B" = "salmon")) +
  theme_minimal()
```

```r
library(ggplot2)

# Función original
f6 <- function(x) ifelse(x >= 0 & x <= 10, (3/14)*(2*x^2 + 2*x + 3), 0)

# Dominio para graficar
x_vals <- seq(0, 10, length.out = 300)

# Evaluar f(x)
fx_vals <- f6(x_vals)

# Calcular F(x): integral acumulada desde 0 hasta x
F_vals <- sapply(x_vals, function(x) integrate(f6, lower = 0, upper = x)$value)

# Crear data frame para ggplot
df <- data.frame(x = x_vals, fx = fx_vals, Fx = F_vals)

# Graficar f(x) y su integral acumulada F(x)
ggplot(df, aes(x = x)) +
  geom_line(aes(y = fx), color = "blue", size = 1.2, linetype = "solid") +
  geom_line(aes(y = Fx), color = "red", size = 1.2, linetype = "dashed") +
  labs(title = "Función f(x) y su integral acumulada F(x)",
       subtitle = expression(f(x) == (3/14)*(2*x^2 + 2*x + 3)~y~F(x) == integral(f(t)*dt, 0, x)),
       x = "x",
       y = "Valor") +
  theme_minimal() +
  scale_y_continuous(sec.axis = sec_axis(~., name = "F(x) — Área acumulada")) +
  theme(legend.position = "none")
```
